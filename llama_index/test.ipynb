```python
from dotenv import load_dotenv
import pandas as pd
import os
from llama_index.experimental.query_engine import PandasQueryEngine
from prompts import new_prompt, instruction_str, context
from custom_tools import check_date
from llama_index.core.tools import QueryEngineTool, ToolMetadata
from llama_index.core.agent import ReActAgent
from llama_index.llms.openai import OpenAI

load_dotenv('../keys.env')
```

```python
df = pd.read_csv('../__data/Amazon_Sales_Report.csv')
```

```python
engine = PandasQueryEngine(df=df, verbose=True, instruction_str=instruction_str)
engine.update_prompts({'pandas_prompt':new_prompt})
```

```python
tools = [
    engine,
    QueryEngineTool(
        query_engine=engine,
        metadata=ToolMetadata(
            name='check_date_columns',
            description='This data gives information about sales of Amazon'
        )
    )
]

llm = OpenAI(model='gpt-3.5-turbo-0125')
agent = ReActAgent.from_tools(tools, llm=llm, verbose=True, context=context)
```

```python
agent.query('What are the format of the "Date" columns?')
```
